## 1. Introduction to Matplotlib
**Exercise**: Create a simple line plot using Matplotlib. Plot the function $y = x^2$ for $x$ ranging from -10 to 10. Customize the plot by adding a title, x and y labels, and a grid.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y)
plt.title('Plot of $y = x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## 2. Simple Line Plots
**Exercise**: Plot two functions, $y = \sin(x)$ and $y = \cos(x)$, on the same graph for $x$ ranging from 0 to $2\pi$. Use different line styles and colors for each function. Add a legend to distinguish between the two lines.

```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)', linestyle='-', color='blue')
plt.plot(x, y2, label='cos(x)', linestyle='--', color='red')
plt.title('Plot of sin(x) and cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## 3. Simple Scatter Plots
**Exercise**: Generate a scatter plot of 100 random points. Use different colors and sizes for the points based on their x and y values. Add labels to the axes and a title to the plot.

```python
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.title('Random Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
```

## 4. Errorbars
**Exercise**: Create a plot with error bars. Use the function $y = \sin(x)$ with $x$ ranging from 0 to 10. Add random noise to the y-values and plot the error bars to represent the noise. Customize the appearance of the error bars.

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=0)
plt.title('Errorbar Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## 5. Density and Contour Plots
**Exercise**: Generate a contour plot for the function $z = \sin(x^2 + y^2)$ over the range $-5 \leq x, y \leq 5$. Use a color map to represent different levels of the function.

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar()
plt.title('Contour Plot of $z = \sin(x^2 + y^2)$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## 6. Histograms and Binnings
**Exercise**: Create a histogram of 1000 random numbers drawn from a normal distribution. Experiment with different numbers of bins and observe how the histogram changes. Add a title and labels to the axes.

## 7. Customizing Legends
**Exercise**: Plot two functions, $y = e^x$ and $y = e^{-x}$, on the same graph. Customize the legend to include a title and place it outside the plot area.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

x = np.linspace(0, 2, 100)
y1 = np.exp(x)
y2 = np.exp(-x)

plt.plot(x, y1, label='exp(x)', color='green')
plt.plot(x, y2, label='exp(-x)', color='purple')
plt.title('Exponential Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(title='Functions', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

## 8. Customizing Colorbars
**Exercise**: Create a heatmap of a 10x10 matrix of random numbers. Add a colorbar to the plot and customize its appearance by changing its orientation and adding a label.

## 9. Multiple Subplots
**Exercise**: Create a figure with four subplots arranged in a 2x2 grid. Plot different mathematical functions in each subplot. Ensure each subplot has its own title and axis labels.

```python
x = np.linspace(0, 2 * np.pi, 100)

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('sin(x)')

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title('cos(x)')

axs[1, 0].plot(x, np.tan(x))
axs[1, 0].set_title('tan(x)')

axs[1, 1].plot(x, np.log(x + 1))
axs[1, 1].set_title('log(x+1)')

for ax in axs.flat:
    ax.set(xlabel='x', ylabel='y')

plt.tight_layout()
plt.show()
```

## 10. Text and Annotation
**Exercise**: Plot the function $y = \log(x)$ for $x$ ranging from 1 to 10. Add annotations to highlight the points where $x = 2$ and $x = 5$. Include text labels for these points.

```python
x = np.linspace(1, 10, 100)
y = np.log(x)

plt.plot(x, y)
plt.title('Plot of $y = \log(x)$')
plt.xlabel('x')
plt.ylabel('y')

plt.annotate('x=2', xy=(2, np.log(2)), xytext=(3, 1),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('x=5', xy=(5, np.log(5)), xytext=(6, 1.5),
             arrowprops=dict(facecolor='red', shrink=0.05))

plt.show()
```

## 11. Customizing Ticks
**Exercise**: Plot the function $y = \tan(x)$ for $x$ ranging from $-\pi/2$ to $\pi/2$. Customize the ticks on the x-axis to show multiples of $\pi/4$.

```python
x = np.linspace(-np.pi/2, np.pi/2, 100)
y = np.tan(x)

plt.plot(x, y)
plt.title('Plot of $y = \tan(x)$')
plt.xlabel('x')
plt.ylabel('y')

plt.xticks(ticks=[-np.pi/4, 0, np.pi/4], labels=['$-\pi/4$', '0', '$\pi/4$'])
plt.ylim(-10, 10)
plt.grid(True)
plt.show()
```

## 13. Three-Dimensional Plotting
**Exercise**: Create a 3D surface plot of the function $z = \cos(\sqrt{x^2 + y^2})$ over the range $-5 \leq x, y \leq 5$. Use a color map to represent the height of the surface.

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.cos(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('3D Surface Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
